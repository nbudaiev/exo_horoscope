```python
#import pandas as pd
from astroquery.ipac.nexsci.nasa_exoplanet_archive import NasaExoplanetArchive

# Query the NASA Exoplanet Archive for exoplanets
exoplanets_table = NasaExoplanetArchive.query_criteria(table="pscomppars", select="*")

# Create a new table with the star's coordinates and the exoplanet's mass
table = exoplanets_table[['ra', 'dec']]

# Convert the table to a pandas DataFrame
#df = table.to_pandas()

# Print the DataFrame
#print(df)
```

```python
exoplanet_coordinates = SkyCoord(exoplanets_table['ra'], exoplanets_table['dec'], unit=(u.deg, u.deg))
gal = exoplanet_coordinates.galactic
```

```python
exoplanets_table_2016 = exoplanets_table[exoplanets_table['disc_year'] < 2017]
exoplanet_coordinates_2016 = SkyCoord(exoplanets_table_2016['ra'], exoplanets_table_2016['dec'], unit=(u.deg, u.deg))
gal_2016  = exoplanet_coordinates_2016.galactic
```

```python
plt.figure(figsize=(6,5))
fig = plt.figure()
ax = fig.add_subplot(111, projection="aitoff")
#plt.plot(gal.l.wrap_at(180*u.deg), gal.b.wrap_at(180*u.deg), linestyle='None', label ='2024')
#plt.plot(gal_2016.l.wrap_at(180*u.deg), gal_2016.b.wrap_at(180*u.deg), linestyle='None', label ='2016')
plt.grid(True)
ax.scatter(gal.l.wrap_at('180d').radian, gal.b.radian, linestyle='None', marker='o', s=3, label = '2024', color='darkviolet', alpha = 0.6)
ax.scatter(gal_2016.l.wrap_at('180d').radian, gal_2016.b.radian, linestyle='None', marker='o', s=3, label = '2016', color ='darkgreen', alpha = 0.6)
plt.legend()

#plt.scatter(gal.l.wrap_at('180d').radian, gal.b.radian)
plt.savefig('galactic_map.png', dpi=300)
```
